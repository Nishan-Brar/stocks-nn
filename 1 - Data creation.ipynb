## Numerical features

```python
first_date = stocks_df.index[0]
last_date = stocks_df.index[-1]
```

```python
leading_indicators = pd.read_csv(f"https://stats.oecd.org/SDMX-JSON/data/MEI_CLI/LOLITOAA+BSCICP03+CSCICP03.USA.M/all?startTime={first_date.year}-0{first_date.month}&endTime={last_date.year}-{last_date.month}&dimensionAtObservation=allDimensions&contentType=csv")

remap = {"BSCICP03": "BCI", # Business confidence indicator (BCI)
         "CSCICP03":"CCI", # Comsumer confidence indicator (CCI)
         "LOLITOAA":"CLI"} # Composite leading indicator (CLI)
leading_indicators.replace({"SUBJECT": remap},inplace=True)

leading_indicators = leading_indicators.drop(columns=["Subject",
                                                      "LOCATION",
                                                      "Country",
                                                      "FREQUENCY",
                                                      "Frequency",
                                                      "Time",
                                                      "Unit Code",
                                                      "Unit",
                                                      "PowerCode Code",
                                                      "PowerCode",
                                                      "Reference Period Code",
                                                      "Reference Period",
                                                      "Flag Codes",
                                                      "Flags"]
                                             ).set_index("TIME")

leading_indicators = leading_indicators.pivot(columns="SUBJECT", values="Value")
leading_indicators.index = pd.to_datetime(leading_indicators.index)
leading_indicators
```

```python
more_indicators = pd.read_csv(f"https://stats.oecd.org/SDMX-JSON/data/KEI/PRINTO01+PRMNTO01+PRCNTO01+SLRTTO01+SLRTCR03+LOLITOAA+BSCICP02+LFEMTTTT+LCEAMN01+PIEAMP01+CPALTT01+MANMM101+MABMM301+IR+IRSTCI01+IR3TIB01+IRLTLT01.USA.ST.M/all?startTime={first_date.year}-0{first_date.month}&endTime={last_date.year}-{last_date.month}&dimensionAtObservation=allDimensions&contentType=csv")

more_indicators = more_indicators.drop(columns=["SUBJECT",
                                                "LOCATION",
                                                "Country",
                                                "MEASURE",
                                                "Measure",
                                                "FREQUENCY",
                                                "Frequency",
                                                "Time",
                                                "Unit Code",
                                                "Unit",
                                                "PowerCode Code",
                                                "PowerCode",
                                                "Reference Period Code",
                                                "Reference Period",
                                                "Flag Codes",
                                                "Flags"]
                                       ).set_index("TIME")

more_indicators = more_indicators.pivot(columns="Subject", values="Value")
more_indicators.index = pd.to_datetime(more_indicators.index)

more_indicators.columns = ["3 month interbank rate",
                           "Broad money",
                           "Construction",
                           "Consumer prices",
                           "Manufacturing hourly earnings",
                           "Industrial production",
                           "Leading indicator",
                           "Long-term interest rate", # Long-term interest rates refer to government bonds maturing in ten years
                           "Manufacturing confidence indicator",
                           "Narrow money",
                           "Overnight interbank rate",
                           "Car registrations",
                           "Manufacturing producer prices",
                           "Retail trade volume",
                           "Total employment",
                           "Total manufacturing"]

more_indicators
```

## Combining numerical features into a daily-dataframe

Using "daily" since stock prices are given with daily fequency.

```python
features_monthly = pd.concat([leading_indicators,
                              more_indicators], axis = "columns")

features_monthly.loc[stocks_df.index[-1],:] = [np.nan] * len(features_monthly.columns) # This is to ensure that every day in the stocks_df has a corresponding column
features_daily = features_monthly.resample("D").nearest()
# features_daily = pd.concat([features_daily, yield_curve], axis = "columns") ######################## why is this hash-tagged-out?? is it only if I use yield curve?? try with hashed-out first

# Filter so that the index of this df matches up with the index of `stocks_df` so you can convert into numpy array
features_daily = features_daily.loc[stocks_df.index,:]

features_daily = features_daily.drop(columns = ["Leading indicator"])

features_daily = features_daily.fillna(method="ffill")
features_daily.to_csv("data/features.csv")

features_daily
```
